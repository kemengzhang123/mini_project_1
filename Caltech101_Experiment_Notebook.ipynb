## 1. End-to-End Experiment Execution

The following cell contains the full training and evaluation pipeline for all four ablation settings:

- `S1`: 64 no augmentation
- `S2`: 64 augmentation
- `S3`: 128 no augmentation
- `S4`: 128 augmentation

It includes data preparation, model training, evaluation metrics, and saving of result artifacts.


```python
# ============================================================
# Mini Project 1 - Caltech101 Classification
# Models:
#   1) KNN
#   2) ResNet18
#   3) EfficientNet-B0
# Ablation:
#   - image size: 64 vs 128
#   - augmentation: off vs on
# Split:
#   - 70 / 15 / 15
# ============================================================

# =========================
# 1. Mount Drive
# =========================
from google.colab import drive
drive.mount('/content/drive')

# =========================
# 2. Imports
# =========================
import os
import gc
import json
import time
import copy
import glob
import shutil
import random
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm.auto import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, ConcatDataset
from torchvision import datasets, transforms, models

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    f1_score
)

# !pip install kagglehub
import kagglehub

# =========================
# 3. Config
# =========================
class CFG:
    seed = 3407
    exp_root = "/content/drive/MyDrive/mp1_caltech101_no_vit_runs"
    local_data_root = "/content/caltech101_split_no_vit"

    batch_size = 16
    num_workers = 0      # Colab-safe
    epochs = 10
    lr = 3e-4
    weight_decay = 1e-4

    knn_neighbors = 5

    run_knn = True
    run_resnet18 = True
    run_efficientnet = True

    ablations = [
        {"name": "S1_64_noaug",  "image_size": 64,  "use_aug": False},
        {"name": "S2_64_aug",    "image_size": 64,  "use_aug": True},
        {"name": "S3_128_noaug", "image_size": 128, "use_aug": False},
        {"name": "S4_128_aug",   "image_size": 128, "use_aug": True},
    ]

os.makedirs(CFG.exp_root, exist_ok=True)
os.makedirs(CFG.local_data_root, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# =========================
# 4. Reproducibility
# =========================
def set_seed(seed: int):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(CFG.seed)

# =========================
# 5. Dataset Preparation
# =========================
def prepare_caltech101_split(output_root: str):
    train_dir = os.path.join(output_root, "train")
    val_dir   = os.path.join(output_root, "val")
    test_dir  = os.path.join(output_root, "test")

    if all(os.path.exists(p) and len(os.listdir(p)) > 0 for p in [train_dir, val_dir, test_dir]):
        print("Dataset split already exists. Skipping preparation.")
        return

    dataset_path = kagglehub.dataset_download("imbikramsaha/caltech-101")
    print("Downloaded dataset path:", dataset_path)

    root = "/kaggle/input/caltech-101/caltech-101"
    if not os.path.exists(root):
        maybe = os.path.join(dataset_path, "caltech-101")
        if os.path.exists(maybe):
            root = maybe
        else:
            root = dataset_path

    print("Using dataset root:", root)

    for split_name in ["train", "val", "test"]:
        os.makedirs(os.path.join(output_root, split_name), exist_ok=True)

    class_names = sorted([
        d for d in os.listdir(root)
        if os.path.isdir(os.path.join(root, d))
        and d != "BACKGROUND_Google"
        and not d.startswith(".")
    ])

    print(f"Found {len(class_names)} valid classes")

    for cls in tqdm(class_names, desc="Creating split"):
        cls_path = os.path.join(root, cls)
        imgs = sorted(glob.glob(os.path.join(cls_path, "*.jpg")))

        if len(imgs) < 3:
            continue

        train_imgs, temp_imgs = train_test_split(
            imgs, test_size=0.30, random_state=CFG.seed
        )
        val_imgs, test_imgs = train_test_split(
            temp_imgs, test_size=0.50, random_state=CFG.seed
        )

        split_map = {
            "train": train_imgs,
            "val": val_imgs,
            "test": test_imgs
        }

        for split_name, files in split_map.items():
            dst_dir = os.path.join(output_root, split_name, cls)
            os.makedirs(dst_dir, exist_ok=True)
            for f in files:
                shutil.copy(f, dst_dir)

    print("Dataset split complete.")

prepare_caltech101_split(CFG.local_data_root)

# =========================
# 6. Transforms
# =========================
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

def get_transforms(image_size: int, use_aug: bool):
    train_list = [transforms.Resize((image_size, image_size))]

    if use_aug:
        train_list += [
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomRotation(12),
            transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.10),
        ]

    train_list += [
        transforms.ToTensor(),
        transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
    ]

    eval_tfms = transforms.Compose([
        transforms.Resize((image_size, image_size)),
        transforms.ToTensor(),
        transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
    ])

    return transforms.Compose(train_list), eval_tfms

# =========================
# 7. Dataloaders
# =========================
def build_dataloaders(image_size: int, use_aug: bool):
    train_tfms, eval_tfms = get_transforms(image_size, use_aug)

    train_ds = datasets.ImageFolder(os.path.join(CFG.local_data_root, "train"), transform=train_tfms)
    val_ds   = datasets.ImageFolder(os.path.join(CFG.local_data_root, "val"), transform=eval_tfms)
    test_ds  = datasets.ImageFolder(os.path.join(CFG.local_data_root, "test"), transform=eval_tfms)

    train_loader = DataLoader(
        train_ds,
        batch_size=CFG.batch_size,
        shuffle=True,
        num_workers=0,
        pin_memory=False
    )

    val_loader = DataLoader(
        val_ds,
        batch_size=CFG.batch_size,
        shuffle=False,
        num_workers=0,
        pin_memory=False
    )

    test_loader = DataLoader(
        test_ds,
        batch_size=CFG.batch_size,
        shuffle=False,
        num_workers=0,
        pin_memory=False
    )

    return train_ds, val_ds, test_ds, train_loader, val_loader, test_loader

# =========================
# 8. Models
# =========================
def create_model(model_name: str, num_classes: int):
    if model_name == "resnet18":
        model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        in_features = model.fc.in_features
        model.fc = nn.Linear(in_features, num_classes)

    elif model_name == "efficientnet_b0":
        model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
        in_features = model.classifier[1].in_features
        model.classifier[1] = nn.Linear(in_features, num_classes)

    else:
        raise ValueError(f"Unsupported model: {model_name}")

    return model.to(device)

# =========================
# 9. Metrics
# =========================
def topk_accuracy(probs: np.ndarray, y_true: np.ndarray, k: int = 5):
    topk = np.argsort(probs, axis=1)[:, ::-1][:, :k]
    correct = 0
    for i in range(len(y_true)):
        if y_true[i] in topk[i]:
            correct += 1
    return correct / len(y_true)

def compute_metrics(y_true, y_pred, y_prob=None):
    metrics = {
        "accuracy": accuracy_score(y_true, y_pred),
        "macro_f1": f1_score(y_true, y_pred, average="macro"),
        "weighted_f1": f1_score(y_true, y_pred, average="weighted"),
    }
    if y_prob is not None:
        metrics["top5"] = topk_accuracy(y_prob, np.array(y_true), k=5)
    return metrics

# =========================
# 10. Train / Eval Loops
# =========================
def train_one_epoch(model, loader, criterion, optimizer):
    model.train()
    total_loss = 0.0

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        logits = model(images)
        loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * images.size(0)

    return total_loss / len(loader.dataset)

@torch.no_grad()
def evaluate_epoch(model, loader, criterion):
    model.eval()

    total_loss = 0.0
    all_labels = []
    all_preds = []
    all_probs = []

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        logits = model(images)
        loss = criterion(logits, labels)

        probs = torch.softmax(logits, dim=1)
        preds = torch.argmax(probs, dim=1)

        total_loss += loss.item() * images.size(0)

        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(preds.cpu().numpy())
        all_probs.append(probs.cpu())

    all_probs = torch.cat(all_probs).numpy()
    avg_loss = total_loss / len(loader.dataset)

    metrics = compute_metrics(all_labels, all_preds, all_probs)
    metrics["loss"] = avg_loss

    return metrics, np.array(all_labels), np.array(all_preds), all_probs

def train_model_full(model_name, train_loader, val_loader, num_classes, save_dir):
    model = create_model(model_name, num_classes)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(
        model.parameters(),
        lr=CFG.lr,
        weight_decay=CFG.weight_decay
    )
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=CFG.epochs)

    best_state = None
    best_val_acc = -1
    history = []

    for epoch in range(1, CFG.epochs + 1):
        start_time = time.time()

        train_loss = train_one_epoch(model, train_loader, criterion, optimizer)
        val_metrics, _, _, _ = evaluate_epoch(model, val_loader, criterion)

        scheduler.step()

        row = {
            "epoch": epoch,
            "train_loss": train_loss,
            "val_loss": val_metrics["loss"],
            "val_acc": val_metrics["accuracy"],
            "val_macro_f1": val_metrics["macro_f1"],
            "val_weighted_f1": val_metrics["weighted_f1"],
            "val_top5": val_metrics["top5"],
            "lr": optimizer.param_groups[0]["lr"],
            "epoch_time_sec": time.time() - start_time
        }
        history.append(row)

        print(
            f"[{model_name}] Epoch {epoch:02d}/{CFG.epochs} | "
            f"train_loss={train_loss:.4f} | "
            f"val_loss={val_metrics['loss']:.4f} | "
            f"val_acc={val_metrics['accuracy']:.4f} | "
            f"val_top5={val_metrics['top5']:.4f}"
        )

        if val_metrics["accuracy"] > best_val_acc:
            best_val_acc = val_metrics["accuracy"]
            best_state = copy.deepcopy(model.state_dict())

    os.makedirs(save_dir, exist_ok=True)

    best_path = os.path.join(save_dir, f"{model_name}_best.pth")
    torch.save(best_state, best_path)

    hist_df = pd.DataFrame(history)
    hist_df.to_csv(os.path.join(save_dir, f"{model_name}_history.csv"), index=False)

    model.load_state_dict(best_state)
    return model, hist_df, best_path

@torch.no_grad()
def evaluate_on_test(model, test_loader, class_names, save_dir, model_name):
    criterion = nn.CrossEntropyLoss()
    metrics, y_true, y_pred, y_prob = evaluate_epoch(model, test_loader, criterion)

    print(f"\n===== Test Metrics: {model_name} =====")
    print(json.dumps(metrics, indent=2))

    report_dict = classification_report(
        y_true, y_pred, target_names=class_names, digits=4, output_dict=True
    )
    pd.DataFrame(report_dict).transpose().to_csv(
        os.path.join(save_dir, f"{model_name}_classification_report.csv")
    )

    cm = confusion_matrix(y_true, y_pred)
    pd.DataFrame(cm).to_csv(
        os.path.join(save_dir, f"{model_name}_confusion_matrix.csv"),
        index=False
    )

    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation="nearest")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.colorbar()
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

    return metrics

# =========================
# 11. KNN Pipeline
# =========================
def build_knn_datasets(image_size: int):
    tfm = transforms.Compose([
        transforms.Resize((image_size, image_size)),
        transforms.ToTensor()
    ])

    train_ds = datasets.ImageFolder(os.path.join(CFG.local_data_root, "train"), transform=tfm)
    val_ds   = datasets.ImageFolder(os.path.join(CFG.local_data_root, "val"), transform=tfm)
    test_ds  = datasets.ImageFolder(os.path.join(CFG.local_data_root, "test"), transform=tfm)

    full_train = ConcatDataset([train_ds, val_ds])
    return full_train, test_ds, train_ds.classes

def dataset_to_numpy(dataset):
    X, y = [], []
    for img, label in tqdm(dataset, desc="Converting to NumPy"):
        X.append(img.numpy().reshape(-1))
        y.append(label)
    return np.array(X), np.array(y)

def run_knn_experiment(exp_name: str, image_size: int, save_dir: str):
    full_train, test_ds, class_names = build_knn_datasets(image_size=image_size)

    X_train, y_train = dataset_to_numpy(full_train)
    X_test, y_test = dataset_to_numpy(test_ds)

    print("KNN train shape:", X_train.shape)
    print("KNN test shape :", X_test.shape)

    knn = KNeighborsClassifier(
        n_neighbors=CFG.knn_neighbors,
        weights="distance",
        n_jobs=-1
    )

    start_time = time.time()
    knn.fit(X_train, y_train)
    train_time = time.time() - start_time

    y_pred = knn.predict(X_test)
    y_prob = knn.predict_proba(X_test) if hasattr(knn, "predict_proba") else None

    metrics = compute_metrics(y_test, y_pred, y_prob)
    metrics["train_time_sec"] = train_time

    os.makedirs(save_dir, exist_ok=True)

    report_dict = classification_report(
        y_test, y_pred, target_names=class_names, digits=4, output_dict=True
    )
    pd.DataFrame(report_dict).transpose().to_csv(
        os.path.join(save_dir, "knn_classification_report.csv")
    )

    cm = confusion_matrix(y_test, y_pred)
    pd.DataFrame(cm).to_csv(
        os.path.join(save_dir, "knn_confusion_matrix.csv"),
        index=False
    )

    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation="nearest")
    plt.title(f"Confusion Matrix - KNN ({exp_name})")
    plt.colorbar()
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

    print("\n===== Test Metrics: KNN =====")
    print(json.dumps(metrics, indent=2))

    return metrics

# =========================
# 12. Plot History
# =========================
def plot_training_history(hist_df: pd.DataFrame, title_prefix=""):
    plt.figure(figsize=(6, 4))
    plt.plot(hist_df["epoch"], hist_df["train_loss"], label="train_loss")
    plt.plot(hist_df["epoch"], hist_df["val_loss"], label="val_loss")
    plt.title(f"{title_prefix} Loss Curve")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 4))
    plt.plot(hist_df["epoch"], hist_df["val_acc"], label="val_acc")
    plt.title(f"{title_prefix} Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.tight_layout()
    plt.show()

# =========================
# 13. Main Experiment Loop
# =========================
all_results = []

for ab in CFG.ablations:
    exp_name = ab["name"]
    image_size = ab["image_size"]
    use_aug = ab["use_aug"]

    print("\n" + "=" * 70)
    print(f"Running experiment: {exp_name}")
    print(f"image_size = {image_size}, use_aug = {use_aug}")
    print("=" * 70)

    exp_dir = os.path.join(CFG.exp_root, exp_name)
    os.makedirs(exp_dir, exist_ok=True)

    # ----- KNN -----
    if CFG.run_knn:
        knn_dir = os.path.join(exp_dir, "knn")
        knn_metrics = run_knn_experiment(
            exp_name=exp_name,
            image_size=image_size,
            save_dir=knn_dir
        )

        all_results.append({
            "experiment": exp_name,
            "model": "KNN",
            "image_size": image_size,
            "augmentation": use_aug,
            "accuracy": knn_metrics["accuracy"],
            "macro_f1": knn_metrics["macro_f1"],
            "weighted_f1": knn_metrics["weighted_f1"],
            "top5": knn_metrics.get("top5", np.nan)
        })

    # ----- Deep Learning -----
    train_ds, val_ds, test_ds, train_loader, val_loader, test_loader = build_dataloaders(
        image_size=image_size,
        use_aug=use_aug
    )

    class_names = train_ds.classes
    num_classes = len(class_names)

    model_list = []
    if CFG.run_resnet18:
        model_list.append("resnet18")
    if CFG.run_efficientnet:
        model_list.append("efficientnet_b0")

    for model_name in model_list:
        print(f"\n----- Training {model_name} for {exp_name} -----")
        save_dir = os.path.join(exp_dir, model_name)

        model, hist_df, best_path = train_model_full(
            model_name=model_name,
            train_loader=train_loader,
            val_loader=val_loader,
            num_classes=num_classes,
            save_dir=save_dir
        )

        plot_training_history(hist_df, title_prefix=f"{model_name} | {exp_name}")

        test_metrics = evaluate_on_test(
            model=model,
            test_loader=test_loader,
            class_names=class_names,
            save_dir=save_dir,
            model_name=model_name
        )

        all_results.append({
            "experiment": exp_name,
            "model": model_name,
            "image_size": image_size,
            "augmentation": use_aug,
            "accuracy": test_metrics["accuracy"],
            "macro_f1": test_metrics["macro_f1"],
            "weighted_f1": test_metrics["weighted_f1"],
            "top5": test_metrics["top5"]
        })

        del model
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

# =========================
# 14. Save Summary Results
# =========================
results_df = pd.DataFrame(all_results)
results_csv = os.path.join(CFG.exp_root, "all_experiment_results.csv")
results_df.to_csv(results_csv, index=False)

print("\nSaved results to:", results_csv)
display(results_df.sort_values(["model", "experiment"]).reset_index(drop=True))

# =========================
# 15. Summary Plots
# =========================
plt.figure(figsize=(10, 5))
for model_name in results_df["model"].unique():
    tmp = results_df[results_df["model"] == model_name].copy()
    tmp = tmp.sort_values("experiment")
    plt.plot(tmp["experiment"], tmp["accuracy"], marker="o", label=model_name)

plt.title("Accuracy Across Experiments")
plt.xlabel("Experiment")
plt.ylabel("Accuracy")
plt.ylim(0, 1.0)
plt.legend()
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
for model_name in results_df["model"].unique():
    tmp = results_df[results_df["model"] == model_name].copy()
    tmp = tmp.sort_values("experiment")
    plt.plot(tmp["experiment"], tmp["macro_f1"], marker="o", label=model_name)

plt.title("Macro F1 Across Experiments")
plt.xlabel("Experiment")
plt.ylabel("Macro F1")
plt.ylim(0, 1.0)
plt.legend()
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

print("\n===== Best configuration per model =====")
best_df = (
    results_df.sort_values(["model", "accuracy"], ascending=[True, False])
    .groupby("model")
    .head(1)
    .reset_index(drop=True)
)
display(best_df)
```

```python
# ============================================================
# Cell 3. Training curves for the best deep learning models
# What this cell does:
# 1) Load the saved history CSV for ResNet18 and EfficientNet-B0
# 2) Plot training loss and validation loss
# 3) Plot validation accuracy
# Why this matters:
# - These plots help explain convergence behavior
# - They are useful for the Discussion section of the report
# ============================================================

import os
import pandas as pd
import matplotlib.pyplot as plt

# Safety check for CFG
if "CFG" not in globals():
    raise ValueError("CFG is not found. Please run the setup/config cell first.")

# Based on your final best configuration screenshot:
# - ResNet18 best: S3_128_noaug
# - EfficientNet-B0 best: S3_128_noaug
history_files = {
    "resnet18": os.path.join(CFG.exp_root, "S3_128_noaug", "resnet18", "resnet18_history.csv"),
    "efficientnet_b0": os.path.join(CFG.exp_root, "S3_128_noaug", "efficientnet_b0", "efficientnet_b0_history.csv"),
}

for model_name, path in history_files.items():
    print(f"\nChecking history file for {model_name}:")
    print(path)

    if not os.path.exists(path):
        print("History file not found. Skip this model.")
        continue

    hist = pd.read_csv(path)

    # -----------------------------
    # Plot 1: Train loss vs Val loss
    # -----------------------------
    # This plot shows whether the model is learning steadily
    # and whether there are signs of overfitting
    plt.figure(figsize=(6, 4))
    plt.plot(hist["epoch"], hist["train_loss"], marker="o", label="Train Loss")
    plt.plot(hist["epoch"], hist["val_loss"], marker="o", label="Val Loss")
    plt.title(f"{model_name} Loss Curve")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # -----------------------------
    # Plot 2: Validation accuracy
    # -----------------------------
    # This plot shows when the model reaches its best validation performance
    plt.figure(figsize=(6, 4))
    plt.plot(hist["epoch"], hist["val_acc"], marker="o", label="Validation Accuracy")
    plt.title(f"{model_name} Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

**Figure 2. Training curves for deep models (best setting visualization).**

These curves show train/validation loss and validation accuracy trends for ResNet-18 and EfficientNet-B0, illustrating stable convergence within the 10-epoch training budget.


```python
# ============================================================
# Cell 4. Top-1 vs Top-5 accuracy
# What this cell does:
# 1) Plot top-1 accuracy (regular accuracy) and top-5 accuracy
# 2) Compare how much performance improves when using top-5 predictions
# Why this matters:
# - This is a nice bonus figure for the report
# - It shows that even when the top prediction is wrong,
#   the true class is often among the top few predictions
# ============================================================

import matplotlib.pyplot as plt
import pandas as pd

# Safety check
if "results_df" not in globals():
    raise ValueError("results_df is not found. Please run the experiment cell first.")

order = ["S1_64_noaug", "S2_64_aug", "S3_128_noaug", "S4_128_aug"]

# We only plot deep learning models here because KNN top-5 is less important for report discussion
deep_df = results_df[results_df["model"].isin(["resnet18", "efficientnet_b0"])].copy()
deep_df["experiment"] = pd.Categorical(deep_df["experiment"], categories=order, ordered=True)
deep_df = deep_df.sort_values(["model", "experiment"])

plt.figure(figsize=(8, 5))

for model_name in deep_df["model"].unique():
    tmp = deep_df[deep_df["model"] == model_name]
    plt.plot(tmp["experiment"], tmp["accuracy"], marker="o", label=f"{model_name} Top-1")
    plt.plot(tmp["experiment"], tmp["top5"], marker="o", linestyle="--", label=f"{model_name} Top-5")

plt.title("Top-1 vs Top-5 Accuracy")
plt.xlabel("Configuration")
plt.ylabel("Accuracy")
plt.ylim(0.7, 1.0)
plt.legend()
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```

**Figure 3. Top-1 vs Top-5 accuracy.**

Top-5 performance is consistently higher than top-1 for deep models, indicating that many errors are near misses where the correct class remains among top-ranked predictions.
